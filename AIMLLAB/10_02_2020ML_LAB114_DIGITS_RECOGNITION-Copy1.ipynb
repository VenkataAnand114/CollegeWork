```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(wines.data)
winesstandardized = scaler.transform(wines.data) 

pdstdscaled =pd.DataFrame(winesstandardized)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split( wines.data,
                                                 wines.target,
                                                 test_size= 0.40)
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

```python
from sklearn import linear_model

logistic = linear_model.LogisticRegression( solver='lbfgs',max_iter = 2000,multi_class = 'multinomial')
```

```python
logistic.fit(x_train,y_train)
print( 'score: ' , 100*logistic.score(x_test,y_test),'%')
```

```python
ypred_test = logistic.predict(x_test)
print(ypred_test)
```

```python
for i in range ( len(y_test)):
    if ( y_test[i] != ypred_test[i]):
        print(i,y_test[i],ypred_test[i])
```

```python
print("classes", logistic.classes_)
print("Total feature weights", logistic.coef_.shape)
print(" Total bias weights",logistic.intercept_.shape)
print("Total iterations", logistic.n_iter_)
```

```python
ypred_logproba = logistic.predict_log_proba(x_test[0:2])
print(ypred_logproba)
ypred_proba  = logistic.predict_proba(x_test[0:2])
print("  ")
print(ypred_proba)
```
